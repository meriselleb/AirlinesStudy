One of which is a data set of the amount of incidents by airline from 1985-1999 and the other being a data set of the amount of incidents by airline from 2000-2014. My null hypothesis for my study is the common belief that airlines that had 'high' amount of incidents from 1985-1999 will also have 'high' amount of incidents from 2000-2014 compared to those who did not have high amounts of incidents in 1985-1999. With this data, I will be able to see if a history of past incidents (the dataset from 1985-1999) has any correlation on the likelihood of future incidents (the dataset from 2000-2014).

    However, while this data would naturally make sense (due to the progression of technology, one would assume newer tech would make newer planes safer than the ones being used in the 1985-1999 period), we still cannot say these statistics properly represent the data sets and actually support that assumption. One reason I am hesitant to do so is because of the suspected outlier in the 1985-1999 dataset of Aeroflot that had an unusally high amount of incidents (76), much higher than the average of 7.18 for that data set. The median (which is resistant to outliers unlike the mean, range, variance, and standard deviation) for the 1985-1999 was 4 while the median for the 2000-2014 data was 3. This is a incredibly small difference. Thus, we cannot be sure if this potential outlier is the reason for why the earlier data set has much higher incident statistics than the more recent data set. So, later on, I will do a Two-Sided Test to see if there truly is a significant difference between the data sets. If there is not significant difference in the data sets, this could go towards supporting our null hypothesis that there is a correlation between the earlier data set and more recent data set.

     I will now create box plots for each of the data sets to see if there are any outliers present that may have had an influence on the statistics previously calculated.

  Looking at the two box plots, we are able to see what outliers each set of data has. As you can see, as suspected, the 76 incidents of Aeroflot is an outlier, and an incredibly far out outlier at that. While both sets do have their own outliers, the earlier data set has more outliers and outliers that are much farther away than the more recent data set. So, as suspected, the outliers may have effected the data and caused it to seem like there is a difference and decrease in the amount of incidents per airline when comparing the two data sets. However, I will see if the data from 1985-1999 and from 2000-2014 are truly signiicantly different with a Two-Sided Test. In this case, we will say the null hypothesis is that there is a significant difference, even though our null hypothesis of the overall study is that there is a correlation and thus would not have a significant difference.  
    If the alternative hypothesis cannot be rejected while the null hypothesis can, then we can continue in testing whether there is a correlation in the distribution and test our overall null hypothesis. 

```python
earlierIncidents = df['incidents_85_99']
recentIncidents = df['incidents_00_14']
meanDiff = np.mean(recentIncidents) - np.mean(earlierIncidents)
print("Difference in Means: ", meanDiff)
```

```python
def custom_pooled_variance(a, b):
    return ((((len(a) - 1) * np.var(a, ddof=1)) + ((len(b) - 1) * np.var(b, ddof=1))) / (len(a) + len(b) - 2)) * ((1/len(a)) + 1/len(b))
pooled_var = custom_pooled_variance(earlierIncidents, recentIncidents)
print('Pooled Variance:', np.round(pooled_var, 2))
print('Pooled Standard Deviation:', np.round(np.sqrt(pooled_var), 2))
```

```python
lengthOf = len(earlierIncidents)
degreesFree = len(earlierIncidents) + len(recentIncidents) -2
print("Degrees of Freedom: ", degreesFree)
dist = stats.t(loc=0,scale=np.sqrt(pooled_var), df = 57)
```

```python
# Two Sided Test at a signifance of 0.05. Rejecting anything before 2.5th percentile and after 97.5th percentile
# (the Two-Tail Critical Regions)
lowerPercentile = dist.ppf(0.025)
upperPercentile = dist.ppf(0.975)
print("Rejection Region Lower Bounds: Anything Less than: ", np.round(lowerPercentile, 2))
print("Rejection Region Upper Bounds: Anything above: ", np.round(upperPercentile,2))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.025), dist.ppf(0.025)],[0, dist.pdf(dist.ppf(0.025))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.0001), dist.ppf(0.025))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.plot([dist.ppf(0.975), dist.ppf(0.975)],[0, dist.pdf(dist.ppf(0.975))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.975), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('Difference in Sample Means')
plt.ylabel('Probability Density Function')
fig.suptitle('T-Distribution of Difference Between Samples', fontsize=15, y=0.92)
```

```python
#P-Value
stats.t(loc=0, scale=np.sqrt(pooled_var), df=degreesFree).cdf(meanDiff) * 2
```

   Since the p-value is 0.05814, the difference in the data is not statistically significant. So, we can not reject the null hypothesis that there is a signifcant difference, and we can not accept the alternative hypothesis that there is no significant difference. So, despite the mean data showing that there was a difference between the two means, after doing the Two Sided Test, there seems to be weak evidence against and for both the null and alternative hypothesis. In this case, due to the outliers, the median was a better representation of the data as the medians of both data sets were relatively similar and did not give significant evidence on whether the data sets were the same or different. 
    However, while this may not support or reject our hypothesis that there is a significant difference between the two data sets, this might help support our main hypothesis stated earlier in that the airlines with a history of incidents in the past will have a higher likelihood of incidents in the future. This is because, if the data sets can not be concluded to be different, they may have the same or similar distribution. If they do, that means there is a correlation between the data sets. That means that we can reasonably expect similar distributions in future data sets. Which, in turn, would be us accepting our main null hypothesis (that airlines with a history of a higher amount of incidents from the earlier data set have a higher amount of incidents in the more recent data set than airlines that did not have a higher amount of incidents in the earlier data set). 
    I originally planned to do a chi-square analysis, but had trouble computing the contingency with 0 values in some of my data. So, I researched on how else I could compare the distribution of two data sets to see if there is enough evidence to declare if their distributions were similar. I came across doing a Kolmogorov-Smirnov statistic. This tests whether 2 data sets are drawn from the same distribution. If the history of an airlines past incidents has any correlation with the likelihood of future incidents, then the two data sets 1985-1999 and 2000-2014 would be drawn from the same distribution.

```python
# Kolmogorov-Smirnov statistic calculation
stats.ks_2samp(df['incidents_85_99'], df['incidents_00_14'])
```

    "When the KS statistic (Kolmogorov-Smirnov statistic) is small or the p-value is high, then we cannot reject the hypothesis of the distribution of the two samples are the same." 
    So, our KS statistic is 0.19 while our p-value is 0.23. This means, relatively, we cannot reject the null hypothesis at a 19% or lower alpha since the p-value at 0.23 is higher than 19%.

While the more recent data set has more outliers, the older data set has a higher mean, larger interquartile range, and overall range not including outliers. I will now do a T distribution test to see if the difference in the data sets is statistically significant.  

```python
earlierFatalities = df['fatalities_85_99']
recentFatalities = df['fatalities_00_14']
meanFatal = np.mean(recentFatalities) - np.mean(earlierFatalities)
print("Difference in Means: ", meanFatal)
```

```python
def custom_pooled_variance(a, b):
    return ((((len(a) - 1) * np.var(a, ddof=1)) + ((len(b) - 1) * np.var(b, ddof=1))) / (len(a) + len(b) - 2)) * ((1/len(a)) + 1/len(b))
pooled_var2 = custom_pooled_variance(earlierFatalities, recentFatalities)
print('Pooled Variance:', np.round(pooled_var, 2))
print('Pooled Standard Deviation:', np.round(np.sqrt(pooled_var), 2))
```

```python
lengthOf = len(earlierFatalities)
degreesFree = len(earlierFatalities) + len(recentFatalities) -2
print("Degrees of Freedom: ", degreesFree)
dist = stats.t(loc=0,scale=np.sqrt(pooled_var2), df = 57)
```

```python
# Two Sided Test at a signifance of 0.05. Rejecting anything before 2.5th percentile and after 97.5th percentile
# (the Two-Tail Critical Regions)
lowerPercentile = dist.ppf(0.025)
upperPercentile = dist.ppf(0.975)
print("Rejection Region Lower Bounds: Anything Less than: ", np.round(lowerPercentile, 2))
print("Rejection Region Upper Bounds: Anything above: ", np.round(upperPercentile,2))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.025), dist.ppf(0.025)],[0, dist.pdf(dist.ppf(0.025))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.0001), dist.ppf(0.025))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.plot([dist.ppf(0.975), dist.ppf(0.975)],[0, dist.pdf(dist.ppf(0.975))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.975), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('Difference in Sample Means')
plt.ylabel('Probability Density Function')
fig.suptitle('T-Distribution of Difference Between Samples', fontsize=15, y=0.92)
```

```python
#P-Value
stats.t(loc=0, scale=np.sqrt(pooled_var2), df=degreesFree).cdf(meanFatal) * 2
```

Since the p-value is 0.0226, the difference is, again, not statistically significant. Thus, we reached a similar result and thus the same conclusions.

Conclusion:
    From the tests done so far, all evidence says we cannot reject the null hypothesis that there is a similar if not the same distribution of number of incidents for the airlines. From the T Distribution test, there is weak evidence in which we cannot reject the null hypothesis, but we can also not reject the alternative hypothesis. From the Kolmovgorov-Smirnov test, it leans more towards the null hypothesis.
